```python
# Convert day, month, year to Python datetime objects
year = []
month = []
day = []
for key, post in posts_dictionary.items():
    year.append(post['year'])
    month.append(post['month'])
    day.append(post['day'])
```

```python
datedf = pd.DataFrame({'day': day, 'month': month, 'year': year})
```

```python
date = []
for index, row in datedf.iterrows():
    datestr = str(row['month'] + '-' + row['day'] + '-' + row['year'])
    dt = parser.parse(datestr)              
    date.append(dt)
```
